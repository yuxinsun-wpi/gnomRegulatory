```python
import pandas as pd
bed_file = 'connect_ENSG00000169174.bed'

# Open the file
with open(bed_file, 'r') as f:
    # Read the header (column names)
    #header = f.readline().strip().split(',')
    header =["chromosome","start","end", "value"]
    
    # Initialize a dictionary with empty lists for each column
    data = {col: [] for col in header}
    
    # Iterate over each line in the file
    for line in f:
        # Split the line into values
        values = line.strip().split()

        if len(values) >= 3:
            # Append values to the corresponding lists in the dictionary
            for i, col in enumerate(header):
                data[col].append(values[i])

# Convert the dictionary to a DataFrame
rf = pd.DataFrame(data)


rf['start'] = rf['start'].astype(int)

rf['end'] = rf['end'].astype(int)

rf
```

```python
import pandas as pd

# Sample data
data1 = gf
df1 = pd.DataFrame(data1)

data2 = rf
df2 = pd.DataFrame(data2)


# List to store merged rows
merged_rows = []

# Loop through each row in df1 and df2
for index1, row1 in df1.iterrows():
    for index2, row2 in df2.iterrows():
        if row2['start'] <= row1['Position'] <= row2['end']:
            # Combine the rows
            combined_row = row1.to_dict()
            combined_row.update(row2.to_dict())
            merged_rows.append(combined_row)

# Convert merged rows to DataFrame
merged_df = pd.DataFrame(merged_rows)

merged_df[['Position', 'start', 'end']]

print(merged_df)
```
