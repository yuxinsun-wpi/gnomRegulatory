### Find the species and database for a single identifier e.g. gene, transcript, protein

```python
import requests
import json
import pandas as pd


# Define the Ensembl Gene ID of the gene you want to retrieve
#gene_id = 'ENSG00000157764'
gene_id = 'ENSG00000169174'

# Construct the Ensembl API URL
api_url = f"https://rest.ensembl.org/lookup/id/{gene_id}?expand=1"


# Send a GET request to the Ensembl API.
response = requests.get(api_url, headers={"Content-Type": "application/json"})

# Check for a successful response.
if response.status_code == 200:
    gene_info = response.json()
    print("Gene Symbol:", gene_info["display_name"])
    print("Gene Description:", gene_info["description"])
else:
    print("Failed to retrieve gene information.")
```

```python
#json.loads(gene_info)

pd.DataFrame(gene_info['Transcript'])
```

### Find the species and database for a symbol in a linked external database

```python
gene = 'PCSK9'

#homo_sapiens

api_url = f"https://rest.ensembl.org/lookup/symbol/human/{gene}?expand=1"

# Send a GET request to the Ensembl API.
response = requests.get(api_url, headers={"Content-Type": "application/json"})

# Check for a successful response.
if response.status_code == 200:
    gene_info = response.json()
    print("Gene Symbol:", gene_info["display_name"])
    print("Gene Description:", gene_info["description"])
else:
    print("Failed to retrieve gene information.")
```

```python
#json.loads(gene_info)

pd.DataFrame(gene_info['Transcript'])
```

### Fetch variant consequences based on a variant identifier

```python
rsId = "rs45448095"

api_url = f"https://rest.ensembl.org/vep/human/id/{rsId}?"
 
response = requests.get(api_url, headers={ "Content-Type" : "application/json"})
 
# Check for a successful response.
if response.status_code == 200:
    gene_info = response.json()
else:
    print("Failed to retrieve gene information.")
```
