## Read gnomAD PCSK9 gene variants data

```python
import csv

filename = 'gnomAD_v2.1.1_ENSG00000169174_2023_09_21_22_52_39.csv'

# Read the CSV file into a DataFrame
gf = pd.read_csv(filename)

# Display the first few rows of the DataFrame
print(gf[['Chromosome', 'Position', 'rsIDs', 'Transcript']].head(20))
```

### Merge AS data with gnomAD PCSK9 gene variants based on the Position

```python
import pandas as pd

# Sample data
data1 = gf # gnomeAD
df1 = pd.DataFrame(data1)

data2 = df # AS-variant data
df2 = pd.DataFrame(data2)


# List to store merged rows
merged_rows = []

# Loop through each row in df1 and df2
for index1, row1 in df1.iterrows():
    for index2, row2 in df2.iterrows():
        if row2['start'] <= row1['Position'] <= row2['end']:
            # Combine the rows
            combined_row = row1.to_dict()
            combined_row.update(row2.to_dict())
            merged_rows.append(combined_row)

# Convert merged rows to DataFrame
merged_df = pd.DataFrame(merged_rows)

if len(merged_df) > 0:
    
    merged_df[['Position', 'start', 'end']]

    print(merged_df[['Chromosome', 'Position', 'rsIDs', 'gene_id', 'gene_type']])
```
